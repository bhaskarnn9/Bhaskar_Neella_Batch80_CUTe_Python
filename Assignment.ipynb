## Given a string for example: “This is the first python test”, get the list of vowels and Given a string for example: “This is the first python test”, get the list of vowels and consonants separately.

```python
input_string = 'This is the first python test'
input_string = input_string.replace(' ', '')
print(input_string)

print(len(input_string))
vowels = []
constants = []
for char in input_string:
    if char in ['a', 'e', 'i', 'o', 'u']:
        vowels.append(char)
    else:
        constants.append(char)

print(vowels)
print(constants)
```

## Given an array, get the kth largest element in the array. Write a function to get this. The inputs to the function are array and the kth value. 
## Hint: Use list.sort() 
## Example: Input arr=[2,4,12,8,6,9,10], k=2 Output: 10

```python
def get_kth_value(arr, k):
    input_array = sorted(arr)
    print(input_array)
    kth_element = input_array[-k]
    return kth_element

get_kth_value([2,4,12,8,6,9,10], 2)
```

## There is an array in which the elements are repeated twice and one element is repeated thrice. Using NumPy, find that repeated element.
## Example:
## Input arr = [2,4,6,2,6,4,8,9,7,7,8,9,8]
## Output = 8

```python
import numpy as np

arr = np.array([2,4,6,2,6,4,8,9,7,7,8,9,8])
uniques, uniq_idx, counts = np.unique(arr,return_index=True,return_counts=True)
duplicates = arr[ uniq_idx[counts>=3]]

print(duplicates)
```

 ## i) Create Numpy arrays for the following points: A(1,2,3) and B(2,4,5).
 ## ii) Multiply the two arrays (element-wise multiplication)
 ## iii) Add the two arrays and divide them by 2. Call it as C (element-wise addition)
 ## iv) Find the distance between the C and the other two points:

```python
# i)
A = np.arange(start=1, stop=4, step=1)
print(A)

B = np.arange(start=2, stop=6, step=1)
B = B[B != 3]
print(B)

# ii)
mult = np.multiply(A, B)
print(mult)

# iii)
C = np.add(A, B)
print(C)

# iv)
dist_c_b =  np.sum(np.square(C-B))
print(dist_c_b)

dist_c_a =  np.sum(np.square(C-A))
print(dist_c_a)
```
